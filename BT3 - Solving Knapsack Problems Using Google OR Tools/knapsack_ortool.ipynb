```python
test_path = 'D:\Document\\University\CS106\BT3 - Solving Knapsack Problems Using Google OR Tools\kplib'
list_testcases = os.listdir(test_path)
list_testcases.remove('.git')
list_testcases.remove('knapsack_ortool.ipynb')
list_testcases.remove('result.csv')
list_sizes = os.listdir(os.path.join(test_path, list_testcases[0]))
list_R = os.listdir(os.path.join(test_path, list_testcases[0], list_sizes[0]))
print("List testcases:", list_testcases)
print("List sizes:", list_sizes)
print("List R:", list_R)
```

```python
TIME_LIMIT = 180 # Thời gian tối đa cho mỗi testcase

def solve(id, path):
    values = []
    weights = [[]]
    capacities = []

    with open(path, 'r') as f:
        f.readline()

        n = int(f.readline())
        capacities = [int(f.readline())]

        f.readline()

        for i in range(n):
            p, w = map(int, f.readline().split())
            values.append(p)
            weights[0].append(w)

    solver.init(values, weights, capacities)
    solver.set_time_limit(TIME_LIMIT)  # Thời gian tối đa cho mỗi testcase
    start = time.time()
    computed_value = solver.solve()

    end = time.time()
    packed_items = []
    packed_weights = []
    total_weight = 0
    # print("Total value =", computed_value)
    for i in range(len(values)):
        if solver.best_solution_contains(i):
            packed_items.append(i)
            packed_weights.append(weights[0][i])
            total_weight += weights[0][i]
    # print("Total weight:", total_weight)
    # print("Packed items:", packed_items)
    # print("Packed_weights:", packed_weights)
    return computed_value, total_weight, packed_items, packed_weights, (end - start)
```

```python
random.seed(23520902)

list_items = []
list_weights = []
with open('result.csv', 'w',newline="") as f:
    writer = csv.writer(f)

    f.write('id,testcase,testname,total value,total weight,optimize\n')
    id = 1

    for testcase in list_testcases:
        for size in list_sizes:
            for R in list_R:
                path = os.path.join(test_path, testcase, size, R)
                list_tests = os.listdir(path)

                pos = random.randint(0, len(list_tests) - 1)

                print(f"id = {id}, Solving {testcase}-{size}-{R}-{list_tests[pos]}...........")
                computed_value, total_weight, packed_items, packed_weights, time_run = solve(id, os.path.join(path, list_tests[pos]))
                print(f"Time run: {time_run:.5f}")
                list_items.append(packed_items)
                list_weights.append(packed_weights)
                writer.writerow([
                    id,
                    testcase,
                    f"{size}-{R}-{list_tests[pos]}",
                    computed_value,
                    total_weight,
                    time_run * 1000 <= TIME_LIMIT * 1000,
                    # str(packed_items),
                    # str(packed_weights)
                ])
                f.flush()
                id += 1
```
